# Part 1

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import random
from itertools import combinations
```

### 1. Task A (network construction):

Network construction<br>
node = username<br>
edge = users belonging to the same group_cols

```python
# =========================
# Task A-1: Network Construction Function
# =========================
def build_graph(df, group_cols):
    graph = nx.Graph()

    usernames = df["username"].unique()

    username_to_id = {}
    id_to_username = {}

    for i, username in enumerate(usernames):
        username_to_id[username] = i
        id_to_username[i] = username
        graph.add_node(i, username=username)

    for _, group in df.groupby(group_cols):
        users = group["username"].unique()

        for u1, u2 in combinations(users, 2):
            node_i = username_to_id[u1]
            node_j = username_to_id[u2]
            # Increment weight for repeated co-participation
            if graph.has_edge(node_i, node_j):
                graph[node_i][node_j]["weight"] += 1
            else:
                graph.add_edge(node_i, node_j, weight=1)

    return graph, username_to_id, id_to_username
```

```python
# =========================
# Task A-2: Dataset Loading and Graph Creation
# =========================

small_df = pd.read_csv("datasets/01.small/-BOT_REQUESTS.csv")
medium_df = pd.read_csv("datasets/02.medium/-ITEMS.csv")
large_df = pd.read_csv("datasets/03.large/-PROPERTY_PROPOSAL.csv")

small_graph, small_username_to_id, small_id_to_username = build_graph(
    small_df,
    ["page_name", "thread_subject"]
)

medium_graph, medium_username_to_id, medium_id_to_username = build_graph(
    medium_df,
    ["page_name", "thread_subject"]
)

large_graph, large_username_to_id, large_id_to_username = build_graph(
    large_df,
    ["page_name", "thread_subject"]
)

print("Small network")
print(small_graph.number_of_nodes(), "nodes,", small_graph.number_of_edges(), "edges")
print(list(small_username_to_id.items())[:10])
print(list(small_id_to_username.items())[:10])

print("Medium network")
print(medium_graph.number_of_nodes(), "nodes,", medium_graph.number_of_edges(), "edges")
print(list(medium_username_to_id.items())[:10])
print(list(medium_id_to_username.items())[:10])

print("Large network")
print(large_graph.number_of_nodes(), "nodes,", large_graph.number_of_edges(), "edges")
print(list(large_username_to_id.items())[:10])
print(list(large_id_to_username.items())[:10])
```

### 2. Task B (network metrics)

```python
# =========================
# Task B-1: Basic Network Statistics
# =========================
def print_basic_stats(graph, name):
    print(name)
    print("Number of nodes:", graph.number_of_nodes())
    print("Number of edges:", graph.number_of_edges())
    print()

print_basic_stats(small_graph, "Small network")
print_basic_stats(medium_graph, "Medium network")
print_basic_stats(large_graph, "Large network")

def draw_network(graph, name, figsize=(8, 6)):
    plt.figure(figsize=figsize)
    pos = nx.spring_layout(graph, seed=42)

    nx.draw(
        graph,
        pos=pos,
        node_size=20,
        node_color="steelblue",
        edge_color="gray",
        width=0.3,
        with_labels=False
    )

    plt.title(name)
    plt.show()

draw_network(small_graph, "Small network")
draw_network(medium_graph, "Medium network")
draw_network(large_graph, "Large network")
```

```python
# =========================
# Task B-2: Structural Metrics Calculation
# =========================
def analyse_graph(graph, name):
    print(name)

    avg_degree = sum(d for _, d in graph.degree()) / graph.number_of_nodes()
    clustering = nx.average_clustering(graph)
    num_components = nx.number_connected_components(graph)

    print("Average degree:", avg_degree)
    print("Clustering coefficient:", clustering)
    print("Connected components:", num_components)

    largest_cc = max(nx.connected_components(graph), key=len)
    G_lcc = graph.subgraph(largest_cc).copy()

    print("Largest connected component size:", G_lcc.number_of_nodes())

    avg_path_length = nx.average_shortest_path_length(G_lcc)
    diameter = nx.diameter(G_lcc)

    print("Average shortest path length:", avg_path_length)
    print("Diameter:", diameter)

    degrees = [d for _, d in graph.degree()]

    plt.figure(figsize=(8, 6))
    plt.hist(degrees, bins=50)
    plt.title("Degree distribution - " + name)
    plt.xlabel("Degree")
    plt.ylabel("Frequency")
    plt.show()

    print()

analyse_graph(small_graph, "Small network")
analyse_graph(medium_graph, "Medium network")
analyse_graph(large_graph, "Large network")
```

```python
# =========================
# Task B-3: Comparison with Random Graphs
# =========================
def compare_random(graph, name):
    n = graph.number_of_nodes()
    m = graph.number_of_edges()

    random_graph = nx.gnm_random_graph(n, m, seed=42)

    real_clustering = nx.average_clustering(graph)
    random_clustering = nx.average_clustering(random_graph)

    largest_cc = max(nx.connected_components(graph), key=len)
    G_lcc = graph.subgraph(largest_cc).copy()

    largest_cc_rand = max(nx.connected_components(random_graph), key=len)
    G_rand_lcc = random_graph.subgraph(largest_cc_rand).copy()

    real_path = nx.average_shortest_path_length(G_lcc)
    random_path = nx.average_shortest_path_length(G_rand_lcc)

    print(name)
    print("Clustering (real):", real_clustering)
    print("Clustering (random):", random_clustering)
    print("Clustering ratio (real/random):", real_clustering / random_clustering)

    print("Path length (real):", real_path)
    print("Path length (random):", random_path)
    print("Path length ratio (real/random):", real_path / random_path)
    print()

compare_random(small_graph, "Small network")
compare_random(medium_graph, "Medium network")
compare_random(large_graph, "Large network")
```

### 3. Task C (epidemic models)

```python
# =========================
# Task C-1: Suspect Selection and Spread Analysis
# =========================
import random
import networkx as nx

def epidemic_check(graph, id_to_username, name, top_k=5):
    editor_1, editor_2 = random.sample(list(graph.nodes()), 2)

    print(name)
    print("Editor 1:", id_to_username[editor_1])
    print("Editor 2:", id_to_username[editor_2])

    dist_1 = nx.single_source_shortest_path_length(graph, editor_1)
    dist_2 = nx.single_source_shortest_path_length(graph, editor_2)

    common_nodes = set(dist_1.keys()).intersection(set(dist_2.keys()))

    if len(common_nodes) > 0:
        priority_list = []

        for node in common_nodes:
            if node == editor_1 or node == editor_2:
                continue

            score = 1 / (dist_1[node] + 1) + 1 / (dist_2[node] + 1)
            priority_list.append((node, score))

        priority_list = sorted(priority_list, key=lambda x: x[1], reverse=True)

        print("Propagation plausibility: plausible within the same connected region")
        print("Top priority editors:")
        for node, score in priority_list[:top_k]:
            print("-", id_to_username[node], "(score =", round(score, 3), ")")

    else:
        print("Propagation plausibility: less plausible because the two editors are in different components")

        priority_1 = []
        for node, d in dist_1.items():
            if node != editor_1:
                score = 1 / (d + 1)
                priority_1.append((node, score))

        priority_2 = []
        for node, d in dist_2.items():
            if node != editor_2:
                score = 1 / (d + 1)
                priority_2.append((node, score))

        priority_1 = sorted(priority_1, key=lambda x: x[1], reverse=True)
        priority_2 = sorted(priority_2, key=lambda x: x[1], reverse=True)

        print("Top priority editors around Editor 1:")
        for node, score in priority_1[:top_k]:
            print("-", id_to_username[node], "(score =", round(score, 3), ")")

        print("Top priority editors around Editor 2:")
        for node, score in priority_2[:top_k]:
            print("-", id_to_username[node], "(score =", round(score, 3), ")")

    print()

epidemic_check(small_graph, small_id_to_username, "Small network")
epidemic_check(medium_graph, medium_id_to_username, "Medium network")
epidemic_check(large_graph, large_id_to_username, "Large network")
```

```python
# =========================
# Task C-2: Summary of Network Properties
# =========================
def summarise_graph(graph, name):
    n = graph.number_of_nodes()
    m = graph.number_of_edges()

    avg_degree = sum(d for _, d in graph.degree()) / n
    clustering = nx.average_clustering(graph)
    components = nx.number_connected_components(graph)

    largest_cc = max(nx.connected_components(graph), key=len)
    G_lcc = graph.subgraph(largest_cc).copy()

    avg_path = nx.average_shortest_path_length(G_lcc)
    diameter = nx.diameter(G_lcc)

    # Random graph (with seed for reproducibility)
    random_graph = nx.gnm_random_graph(n, m, seed=42)
    random_clustering = nx.average_clustering(random_graph)

    largest_cc_rand = max(nx.connected_components(random_graph), key=len)
    G_rand_lcc = random_graph.subgraph(largest_cc_rand).copy()
    random_avg_path = nx.average_shortest_path_length(G_rand_lcc)

    # Ratios (avoid division by zero)
    clustering_ratio = clustering / random_clustering if random_clustering != 0 else None
    path_ratio = avg_path / random_avg_path if random_avg_path != 0 else None

    return {
        "Network": name,
        "Nodes": n,
        "Edges": m,
        "Average degree": avg_degree,
        "Clustering": clustering,
        "Random clustering": random_clustering,
        "Clustering ratio": clustering_ratio,
        "Avg shortest path (LCC)": avg_path,
        "Random avg path (LCC)": random_avg_path,
        "Path ratio": path_ratio,
        "Connected components": components,
        "Diameter (LCC)": diameter
    }

summary_df = pd.DataFrame([
    summarise_graph(small_graph, "Small"),
    summarise_graph(medium_graph, "Medium"),
    summarise_graph(large_graph, "Large")
])

print(summary_df)
```
