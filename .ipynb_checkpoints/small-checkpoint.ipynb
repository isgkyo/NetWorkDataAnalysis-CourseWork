```python
import math
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
from itertools import combinations
from collections import defaultdict
```

```python
df = pd.read_csv("datasets/01.small/-BOT_REQUESTS.csv")

#　Data cleaning
df = df[['page_name', 'thread_subject', 'username']].dropna()
df['page_name'] = df['page_name'].astype(str).str.strip()
df['thread_subject'] = df['thread_subject'].astype(str).str.strip()
df['username'] = df['username'].astype(str).str.strip()
df = df[(df['page_name'] != "") & (df['thread_subject'] != "") & (df['username'] != "")]

G = nx.Graph()

# Group interactions by discussion context
groups = df.groupby(['page_name', 'thread_subject'], sort=False)

for _, group in groups:
    # Unique editors participating in the same discussion
    users = group['username'].unique()
    
    # Add users as nodes
    for user in users:
        G.add_node(user)

    # Connect all co-participating users
    for u1, u2 in combinations(users, 2):
        if G.has_edge(u1, u2):
            G[u1][u2]['weight'] += 1
        else:
            G.add_edge(u1, u2, weight=1)

# Structural summary
print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
print("Connected components:", nx.number_connected_components(G))

isolates = list(nx.isolates(G))
print("Isolates:", len(isolates))

if G.number_of_nodes() > 0:
    lcc = max(nx.connected_components(G), key=len)
    print("Largest CC (nodes):", len(lcc))
```

<h3>Definition</h3>
<ul>
    <li>Nodes represent Wikidata editors, identified by their usernames.</li>
    <li>An undirected edge is created between two editors if they participated in the same thread within the same page.</li>
    <li>The network is modelled as a simple undirected and unweighted graph.</li>
</ul>

<h3>Data Structures</h3>
<ul>
    <li>The raw interaction data were stored in a pandas DataFrame.</li>
    <li>The network was constructed and analysed using a NetworkX Graph object.</li>
    <li>Usernames were used directly as node identifiers.</li>
    <li>Additional information, such as the number of comments made by each editor, can be stored as node attributes.</li>
</ul>

<h3>Algorithmic Approach</h3>
<ul>
    <li>The dataset was grouped by the combination of (page_name, thread_subject).</li>
    <li>For each group, the unique set of participating editors was extracted.</li>
    <li>Each editor was added as a node in the graph.</li>
    <li>All unordered pairs of editors within the same group were connected, forming a clique.</li>
    <li>This procedure was repeated for all discussion contexts in the dataset.</li>
</ul>

```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter

# ---------- LCC ----------
def LCC(G):
    if G.number_of_nodes() == 0:
        return G.copy()
    nodes = max(nx.connected_components(G), key=len)
    return G.subgraph(nodes).copy()

# ---------- Basic stats ----------
def basic_stats(G):
    n, m = G.number_of_nodes(), G.number_of_edges()
    comps = nx.number_connected_components(G) if n else 0
    avg_deg = (2*m/n) if n else np.nan
    density = nx.density(G) if n > 1 else np.nan

    Gl = LCC(G)
    nL = Gl.number_of_nodes()

    C = nx.average_clustering(G) if n > 1 else np.nan
    T = nx.transitivity(G) if n > 2 else np.nan
    CL = nx.average_clustering(Gl) if nL > 1 else np.nan
    TL = nx.transitivity(Gl) if nL > 2 else np.nan

    if nL > 1:
        Lbar = nx.average_shortest_path_length(Gl)
        D = nx.diameter(Gl)
    else:
        Lbar, D = np.nan, np.nan

    return {
        "n": n, "m": m, "components": comps,
        "avg_degree": avg_deg, "density": density,
        "n_LCC": nL, "LCC_frac": (nL/n) if n else np.nan,
        "C_full": C, "T_full": T,
        "C_LCC": CL, "T_LCC": TL,
        "L_LCC": Lbar, "diameter_LCC": D
    }

# ---------- Degree distribution ----------
def plot_degree(G, title="Degree distribution"):
    deg = [d for _, d in G.degree()]
    if not deg:
        print("Empty graph")
        return

    plt.figure()
    plt.hist(deg, bins=30)
    plt.xlabel("Degree"); plt.ylabel("Count"); plt.title(title)
    plt.show()

    c = Counter(deg)
    xs = np.array(sorted(c.keys()))
    ys = np.array([c[k] for k in xs])
    mask = (xs > 0) & (ys > 0)
    xs, ys = xs[mask], ys[mask]

    if len(xs):
        plt.figure()
        plt.scatter(xs, ys)
        plt.xscale("log"); plt.yscale("log")
        plt.xlabel("Degree (log)"); plt.ylabel("Frequency (log)")
        plt.title(title + " (log-log)")
        plt.show()

# ---------- Random comparison (ER) ----------
def compare_ER(G, trials=30, seed=42):
    rng = np.random.default_rng(seed)
    n, m = G.number_of_nodes(), G.number_of_edges()
    if n <= 1:
        return None

    p = (2*m) / (n*(n-1))  # same density
    Gl = LCC(G)
    real = {
        "trial": "REAL",
        "C_LCC": nx.average_clustering(Gl) if Gl.number_of_nodes() > 1 else np.nan,
        "L_LCC": nx.average_shortest_path_length(Gl) if Gl.number_of_nodes() > 1 else np.nan
    }

    rows = [real]
    for t in range(trials):
        s = int(rng.integers(0, 1_000_000))
        Gr = nx.erdos_renyi_graph(n=n, p=p, seed=s)
        Grl = LCC(Gr)
        rows.append({
            "trial": t,
            "C_LCC": nx.average_clustering(Grl) if Grl.number_of_nodes() > 1 else np.nan,
            "L_LCC": nx.average_shortest_path_length(Grl) if Grl.number_of_nodes() > 1 else np.nan
        })

    df = pd.DataFrame(rows)
    C_rand = df[df["trial"] != "REAL"]["C_LCC"].mean()
    L_rand = df[df["trial"] != "REAL"]["L_LCC"].mean()

    C_real = real["C_LCC"]
    L_real = real["L_LCC"]

    gamma = C_real / C_rand if (C_rand and not np.isnan(C_rand)) else np.nan
    lam = L_real / L_rand if (L_rand and not np.isnan(L_rand)) else np.nan
    sigma = gamma / lam if (lam and not np.isnan(lam)) else np.nan

    summary = {
        "C_real": C_real, "L_real": L_real,
        "C_rand_mean": C_rand, "L_rand_mean": L_rand,
        "gamma": gamma, "lambda": lam, "sigma": sigma
    }
    return df, summary

def plot_real_vs_rand(summary):
    labels = ["C (LCC)", "L (LCC)"]
    real_vals = [summary["C_real"], summary["L_real"]]
    rand_vals = [summary["C_rand_mean"], summary["L_rand_mean"]]

    x = np.arange(len(labels))
    w = 0.35
    plt.figure()
    plt.bar(x - w/2, real_vals, w, label="Real")
    plt.bar(x + w/2, rand_vals, w, label="ER mean")
    plt.xticks(x, labels)
    plt.ylabel("Value")
    plt.title("Real vs ER (LCC metrics)")
    plt.legend()
    plt.show()

# ============================
# RUN (Task B)
# ============================

# 1) Basic stats table
stats = basic_stats(G)
stats_df = pd.DataFrame(stats, index=["value"]).T
print(stats_df)

# 2) Degree distribution
plot_degree(G, "Degree distribution (full graph)")

# 3) ER comparison
out = compare_ER(G, trials=30, seed=42)
if out is not None:
    df_cmp, summary = out
    print("\n=== Real vs ER summary ===")
    for k, v in summary.items():
        print(f"{k}: {v}")
    plot_real_vs_rand(summary)

# 4) Component sizes (diagnostic)
comp_sizes = sorted([len(c) for c in nx.connected_components(G)], reverse=True)
plt.figure()
plt.plot(comp_sizes)
plt.xlabel("Component rank")
plt.ylabel("Component size")
plt.title("Connected component sizes")
plt.show()
```

```python
import math
import numpy as np
import pandas as pd
import networkx as nx
from collections import defaultdict

# ---------------------------------------------------------
# Helpers
# ---------------------------------------------------------

def largest_connected_component(G: nx.Graph) -> nx.Graph:
    if G.number_of_nodes() == 0:
        return G.copy()
    largest_cc_nodes = max(nx.connected_components(G), key=len)
    return G.subgraph(largest_cc_nodes).copy()

def _safe_shortest_path_lengths(G: nx.Graph, sources):
    """
    Multi-source shortest path lengths.
    Returns dict: node -> min distance to any source.
    Compatible with older networkx versions.
    """
    if len(sources) == 0 or G.number_of_nodes() == 0:
        return {}

    sources = [s for s in sources if s in G]
    if len(sources) == 0:
        return {}

    # Use BFS manually for multi-source
    dist = {}
    visited = set()
    queue = []

    # Initialize queue with all sources
    for s in sources:
        dist[s] = 0
        visited.add(s)
        queue.append(s)

    # BFS
    while queue:
        u = queue.pop(0)
        for v in G.neighbors(u):
            if v not in visited:
                visited.add(v)
                dist[v] = dist[u] + 1
                queue.append(v)

    return dist

def _edge_weight(G, u, v, default=1.0):
    w = G[u][v].get("weight", default)
    try:
        return float(w)
    except Exception:
        return float(default)

def normalize_series(s: pd.Series) -> pd.Series:
    s = s.astype(float)
    if s.max() == s.min():
        return pd.Series(np.zeros(len(s)), index=s.index)
    return (s - s.min()) / (s.max() - s.min())

# ---------------------------------------------------------
# Task C Part 1:
# "How plausible it is that behaviour has NOT propagated yet"
# ---------------------------------------------------------

def neighborhood_exposure_score(
    G: nx.Graph,
    seeds,
    max_hops: int = 2,
    weight_mode: str = "use",  # "use" or "ignore"
) -> pd.DataFrame:
    """
    For each node, compute exposure from seeds within <= max_hops.
    Exposure is higher if closer to seeds and with higher edge weights.
    Returns dataframe with per-node distance and exposure.
    """
    if G.number_of_nodes() == 0:
        return pd.DataFrame(columns=["node", "dist_to_seed", "exposure"])

    dist = _safe_shortest_path_lengths(G, seeds)  # only nodes reachable get entries
    rows = []
    for v in G.nodes():
        d = dist.get(v, np.inf)
        if d == 0:
            exposure = np.inf  # it's a seed
        elif d <= max_hops:
            # simple distance decay
            exposure = 1.0 / d
        else:
            exposure = 0.0

        rows.append((v, d, exposure))

    df = pd.DataFrame(rows, columns=["node", "dist_to_seed", "exposure"])

    if weight_mode == "use":
        # refine exposure using weighted edges on shortest path tree approximation:
        # for each node within hops, add sum of weights from node to its neighbors that are closer to seeds
        # (cheap proxy for "strong tie" connectivity toward the infected region)
        dist_map = {r[0]: r[1] for r in rows}
        w_boost = []
        for v in G.nodes():
            d = dist_map.get(v, np.inf)
            if not (1 <= d <= max_hops):
                w_boost.append(0.0)
                continue
            boost = 0.0
            for u in G.neighbors(v):
                du = dist_map.get(u, np.inf)
                if du < d:  # moving closer to seed region
                    boost += _edge_weight(G, u, v, default=1.0)
            w_boost.append(boost)
        df["weight_boost"] = w_boost
        # combine: exposure + log(1+boost)
        df["exposure"] = df["exposure"] + np.log1p(df["weight_boost"])

    return df.sort_values(["dist_to_seed", "exposure"], ascending=[True, False])

def not_propagated_plausibility(
    G: nx.Graph,
    seeds,
    tested_nodes,
    max_hops: int = 2,
    threshold_exposure: float = 0.0,
    weight_mode: str = "use",
) -> dict:
    """
    A practical "plausibility it has not propagated yet":
    If tested nodes show high abnormal behaviour, check whether their neighborhood shows high exposure.
    Lower exposure around them => more plausible it hasn't propagated.
    This returns a score in [0,1] (heuristic).
    """
    # Build exposure from seeds to all nodes (who is "at risk" near seeds)
    df = neighborhood_exposure_score(G, seeds=seeds, max_hops=max_hops, weight_mode=weight_mode)
    exposure_map = dict(zip(df["node"], df["exposure"]))

    # Measure how much the tested nodes' neighborhoods appear "exposed"
    # We look at 1-hop and 2-hop around tested nodes
    neigh = set()
    for t in tested_nodes:
        if t not in G:
            continue
        neigh.add(t)
        neigh.update(G.neighbors(t))
        if max_hops >= 2:
            for u in G.neighbors(t):
                neigh.update(G.neighbors(u))

    neigh = list(neigh)
    if len(neigh) == 0:
        return {"plausibility_not_propagated": 1.0, "mean_neighborhood_exposure": 0.0, "n_neigh": 0}

    exposures = np.array([exposure_map.get(v, 0.0) for v in neigh], dtype=float)
    mean_exp = float(np.mean(exposures))

    # Turn mean exposure into plausibility score:
    # higher exposure => less plausible it hasn't propagated
    # scale by a soft threshold
    # p = exp(-alpha * max(0, mean_exp - threshold))
    alpha = 1.0
    x = max(0.0, mean_exp - threshold_exposure)
    plaus = math.exp(-alpha * x)

    return {
        "plausibility_not_propagated": float(plaus),
        "mean_neighborhood_exposure": mean_exp,
        "n_neigh": len(neigh),
        "max_hops": max_hops,
        "threshold_exposure": threshold_exposure,
    }

# ---------------------------------------------------------
# Task C Part 2:
# Priority list of editors to check
# ---------------------------------------------------------

def priority_list_to_check(
    G: nx.Graph,
    seeds,                 # suspected trolling editors (one or two)
    top_k: int = 30,
    max_distance: int = 5,
    use_weights: bool = True,
    use_betweenness: bool = True,
    betweenness_k: int = 200,   # approximation for speed on large graphs
    seed: int = 42,
) -> pd.DataFrame:
    """
    Build a ranked list of users to check next.
    Score combines:
      - proximity to seeds (distance decay)
      - tie strength (sum of weights to closer nodes)
      - centrality (betweenness) as "bridging risk"
    """
    if G.number_of_nodes() == 0:
        return pd.DataFrame(columns=["node", "score", "dist", "risk_proximity", "risk_ties", "risk_bridge"])

    # distances to any seed
    dist = _safe_shortest_path_lengths(G, seeds)
    # include only nodes within max_distance and not the seeds themselves
    candidates = [v for v in G.nodes() if v not in seeds and dist.get(v, np.inf) <= max_distance]
    if len(candidates) == 0:
        return pd.DataFrame(columns=["node", "score", "dist", "risk_proximity", "risk_ties", "risk_bridge"])

    # proximity risk: 1/d
    risk_prox = {v: (1.0 / dist[v]) for v in candidates if dist[v] > 0}

    # ties risk: stronger weighted connections towards seed region
    risk_ties = {}
    for v in candidates:
        d = dist[v]
        tie = 0.0
        for u in G.neighbors(v):
            du = dist.get(u, np.inf)
            if du < d:  # points toward seed region
                tie += (_edge_weight(G, u, v, default=1.0) if use_weights else 1.0)
        risk_ties[v] = math.log1p(tie)

    # bridging risk: betweenness centrality (approx for speed if large)
    risk_bridge = {v: 0.0 for v in candidates}
    if use_betweenness:
        # compute on LCC for stability (optional but common)
        G_lcc = largest_connected_component(G)
        # if huge, approximate by sampling k nodes
        rng = np.random.default_rng(seed)
        if G_lcc.number_of_nodes() > betweenness_k:
            sample = list(rng.choice(list(G_lcc.nodes()), size=betweenness_k, replace=False))
            btw = nx.betweenness_centrality(G_lcc, k=betweenness_k, seed=seed, normalized=True)
        else:
            btw = nx.betweenness_centrality(G_lcc, normalized=True)
        for v in candidates:
            risk_bridge[v] = float(btw.get(v, 0.0))

    df = pd.DataFrame({
        "node": candidates,
        "dist": [dist[v] for v in candidates],
        "risk_proximity": [risk_prox.get(v, 0.0) for v in candidates],
        "risk_ties": [risk_ties.get(v, 0.0) for v in candidates],
        "risk_bridge": [risk_bridge.get(v, 0.0) for v in candidates],
    })

    # normalize components and sum
    df["risk_proximity_n"] = normalize_series(df["risk_proximity"])
    df["risk_ties_n"] = normalize_series(df["risk_ties"])
    df["risk_bridge_n"] = normalize_series(df["risk_bridge"])

    # weights for the final score (tweakable)
    w_prox, w_ties, w_bridge = 0.5, 0.3, 0.2
    df["score"] = (
        w_prox * df["risk_proximity_n"] +
        w_ties * df["risk_ties_n"] +
        w_bridge * df["risk_bridge_n"]
    )

    df = df.sort_values(["score", "dist"], ascending=[False, True]).head(top_k)
    return df[["node", "score", "dist", "risk_proximity", "risk_ties", "risk_bridge"]]

# ---------------------------------------------------------
# Optional: Epidemic/cascade simulation to illustrate spread
# (Independent Cascade model on weighted edges)
# ---------------------------------------------------------

def independent_cascade(
    G: nx.Graph,
    seeds,
    p0: float = 0.05,       # base activation probability
    weight_scale: float = 0.02,  # how much weight increases probability
    max_steps: int = 10,
    seed: int = 0,
) -> dict:
    """
    Independent Cascade: active nodes attempt to activate inactive neighbors once.
    Edge probability: p = 1 - exp(-(p0 + weight_scale * weight))
    Returns: active_sets per step and final active list.
    """
    rng = np.random.default_rng(seed)
    active = set([s for s in seeds if s in G])
    newly_active = set(active)
    steps = [set(active)]

    for _ in range(max_steps):
        next_new = set()
        for u in newly_active:
            for v in G.neighbors(u):
                if v in active:
                    continue
                w = _edge_weight(G, u, v, default=1.0)
                p = 1.0 - math.exp(-(p0 + weight_scale * w))
                if rng.random() < p:
                    next_new.add(v)
        if not next_new:
            break
        active |= next_new
        newly_active = next_new
        steps.append(set(active))

    return {
        "final_active": active,
        "n_final_active": len(active),
        "steps_active_sets": steps,
        "n_steps": len(steps) - 1,
        "params": {"p0": p0, "weight_scale": weight_scale, "max_steps": max_steps},
    }

def cascade_summary_table(cascade_result: dict) -> pd.DataFrame:
    steps = cascade_result["steps_active_sets"]
    sizes = [len(s) for s in steps]
    inc = [sizes[0]] + [sizes[i] - sizes[i-1] for i in range(1, len(sizes))]
    return pd.DataFrame({"step": list(range(len(sizes))), "active_total": sizes, "new_active": inc})

# ---------------------------------------------------------
# Task C runner (example usage)
# ---------------------------------------------------------

# Example: choose two random editors to "test"
# (In the real narrative, the Foundation chooses two randomly each day.)
rng = np.random.default_rng(42)
nodes = list(G.nodes())
tested = list(rng.choice(nodes, size=min(2, len(nodes)), replace=False)) if len(nodes) > 0 else []
print("Tested editors:", tested)

# Suppose the tested editors are flagged as suspicious (seeds)
seeds = tested

# Q1: plausibility that it has NOT propagated to neighbors yet
plaus = not_propagated_plausibility(G, seeds=seeds, tested_nodes=tested, max_hops=2, threshold_exposure=0.0, weight_mode="use")
print("\nPlausibility (not propagated yet):")
for k, v in plaus.items():
    print(f"  {k}: {v}")

# Q2: priority list of who to check next
prio = priority_list_to_check(G, seeds=seeds, top_k=20, max_distance=5, use_weights=True, use_betweenness=True, betweenness_k=200, seed=42)
print("\nTop priority editors to check next:")
print(prio)

# Optional: run a cascade simulation to illustrate how far it might spread
cas = independent_cascade(G, seeds=seeds, p0=0.03, weight_scale=0.015, max_steps=10, seed=1)
print("\nCascade simulation result:")
print("  n_final_active:", cas["n_final_active"])
print("  n_steps:", cas["n_steps"])
print(cascade_summary_table(cas))
```
