```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import random
```

Task01

```python
def build_graph(df, group_cols):
    graph = nx.Graph()

    usernames = df["username"].unique()

    username_to_id = {}
    id_to_username = {}

    for i, username in enumerate(usernames):
        username_to_id[username] = i
        id_to_username[i] = username
        graph.add_node(i)

    for _, group in df.groupby(group_cols):
        users = group["username"].unique()

        for i in range(len(users)):
            for j in range(i + 1, len(users)):
                node_i = username_to_id[users[i]]
                node_j = username_to_id[users[j]]
                graph.add_edge(node_i, node_j)

    return graph, username_to_id, id_to_username
```

```python
small_df = pd.read_csv("datasets/01.small/-BOT_REQUESTS.csv")
medium_df = pd.read_csv("datasets/02.medium/-ITEMS.csv")
large_df = pd.read_csv("datasets/03.large/-PROPERTY_PROPOSAL.csv")
```

```python
df = pd.read_csv("datasets/01.small/-BOT_REQUESTS.csv")
#display(df.head())
print(df.shape)
print(df.columns)
```

```python
df = df[["thread_subject", "username", "page_name"]].dropna().copy()
df = df.drop_duplicates()
#df.head()
print(df.shape)
#print(df.columns)
```

```python
small_graph, small_username_to_id, small_id_to_username = build_graph(
    small_df,
    ["page_name", "thread_subject"]
)

medium_graph, medium_username_to_id, medium_id_to_username = build_graph(
    medium_df,
    ["page_name", "thread_subject"]
)

large_graph, large_username_to_id, large_id_to_username = build_graph(
    large_df,
    ["page_name", "thread_subject"]
)
```

```python
print("Number of nodes: {}".format(
    graph.number_of_nodes()
))
print("Number of edges: {}".format(
    graph.number_of_edges()
))
print("Number of connected components: {}".format(
    nx.number_connected_components(graph)
))

degree_sequence = [deg for _, deg in graph.degree]
print("Average degree: {}".format(
    sum(degree_sequence) / len(degree_sequence)
))
```

```python
print("Number of nodes:", graph.number_of_nodes())
print("Number of edges:", graph.number_of_edges())

print("Average degree:", sum([d for _, d in graph.degree]) / graph.number_of_nodes())

print("Average clustering coefficient:", nx.average_clustering(graph))

print("Number of connected components:", nx.number_connected_components(graph))
```

```python
largest_cc = max(nx.connected_components(graph), key=len)
G_lcc = graph.subgraph(largest_cc)

print("Largest component size:", G_lcc.number_of_nodes())
```

```python
print("Average shortest path length:", nx.average_shortest_path_length(G_lcc))
```

```python
print("Diameter:", nx.diameter(G_lcc))
```

```python
n = graph.number_of_nodes()
m = graph.number_of_edges()

random_graph = nx.gnm_random_graph(n, m)
```

```python
print("Clustering (real):", nx.average_clustering(graph))
print("Clustering (random):", nx.average_clustering(random_graph))
```

```python
largest_cc_rand = max(nx.connected_components(random_graph), key=len)
G_rand_lcc = random_graph.subgraph(largest_cc_rand)

print("Path length (real):", nx.average_shortest_path_length(G_lcc))
print("Path length (random):", nx.average_shortest_path_length(G_rand_lcc))
```

```python
plt.hist([d for _, d in random_graph.degree], bins=50)
plt.title("Random network degree distribution")
plt.show()
```

```python
# choose one suspicious editor randomly
suspicious_editor = random.choice(list(graph.nodes()))

print("Suspicious editor:", suspicious_editor)
print("Username:", id_to_username[suspicious_editor])

# direct neighbours
neighbours = list(graph.neighbors(suspicious_editor))
print("Number of direct neighbours:", len(neighbours))
print("Direct neighbours:", neighbours[:20])

# shortest path lengths from the suspicious editor
distances = nx.single_source_shortest_path_length(graph, suspicious_editor)

# sort by distance
sorted_distances = sorted(distances.items(), key=lambda x: x[1])

print("Closest editors to suspicious editor:")
print(sorted_distances[:20])
```

```python
# choose two suspicious editors randomly
editor_1, editor_2 = random.sample(list(graph.nodes()), 2)

print("Editor 1:", editor_1, id_to_username[editor_1])
print("Editor 2:", editor_2, id_to_username[editor_2])

# shortest path lengths from each editor
dist_1 = nx.single_source_shortest_path_length(graph, editor_1)
dist_2 = nx.single_source_shortest_path_length(graph, editor_2)

# common reachable editors
common_nodes = set(dist_1.keys()).intersection(set(dist_2.keys()))

# priority score = sum of distances to both suspicious editors
priority_list = []
for node in common_nodes:
    score = dist_1[node] + dist_2[node]
    priority_list.append((node, score))

priority_list = sorted(priority_list, key=lambda x: x[1])

print("Priority list (closest to both suspicious editors):")
for node, score in priority_list[:20]:
    print(node, id_to_username[node], score)
```

```python
priority_neighbours_1 = list(graph.neighbors(editor_1))
priority_neighbours_2 = list(graph.neighbors(editor_2))

combined_priority = set(priority_neighbours_1).union(set(priority_neighbours_2))

print("Editors to check first:")
for node in list(combined_priority)[:20]:
    print(node, id_to_username[node])
```
