```python
import pandas as pd

df = pd.read_csv("datasets/task02/dft-road-casualty-statistics-collision-last-5-years.csv")
df.head()
```

```python
center_lat = 53.7997
center_lon = -1.5492
delta = 0.0085

north = center_lat + delta
south = center_lat - delta
east = center_lon + delta
west = center_lon - delta

area = df[
    (df["latitude"].between(south, north)) &
    (df["longitude"].between(west, east))
].copy()

print("Chosen area accident count:", len(area))
print("north =", north)
print("south =", south)
print("east  =", east)
print("west  =", west)
```

```python
import osmnx as ox

bbox = (west, south, east, north)

G = ox.graph_from_bbox(
    bbox,
    network_type="drive"
)

print(G)
```

```python
fig, ax = ox.plot_graph(G)
```

```python
stats = ox.basic_stats(G)
stats
```

```python
stats.keys()
```

```python
import math
import osmnx as ox

# ---- area of chosen bounding box ----
lat_rad = math.radians(center_lat)

meters_per_deg_lat = 111320
meters_per_deg_lon = 111320 * math.cos(lat_rad)

width_m = (east - west) * meters_per_deg_lon
height_m = (north - south) * meters_per_deg_lat

area_m2 = width_m * height_m
area_km2 = area_m2 / 1_000_000

print("width_m:", width_m)
print("height_m:", height_m)
print("area_km2:", area_km2)

# ---- densities ----
node_density = stats["n"] / area_km2
intersection_density = stats["intersection_count"] / area_km2
edge_density = stats["street_length_total"] / area_km2   # metres of street per km^2

print("Average street length:", stats["street_length_avg"])
print("Node density:", node_density)
print("Intersection density:", intersection_density)
print("Edge density:", edge_density)
print("Average circuity:", stats["circuity_avg"])
```

```python
import networkx as nx

G_undirected = ox.convert.to_undirected(G)
largest_cc = max(nx.connected_components(G_undirected), key=len)
G_lcc = G_undirected.subgraph(largest_cc).copy()

diameter = nx.diameter(G_lcc)
print("Spatial diameter:", diameter)
```

```python
fractions = []

for _, row in acc_gdf.iterrows():
    edge_id = row["edge_id"]
    point = row["geometry"]

    edge_info = edge_lookup[edge_id]
    line = edge_info["geometry"]
    road_length = edge_info["length"]

    # 線の始点から事故点までの距離
    dist_from_start = line.project(point)

    # 終点から見た距離
    dist_from_end = road_length - dist_from_start

    # nearest intersection 側までの距離
    nearest_end_dist = min(dist_from_start, dist_from_end)

    # 道路長に対する比率
    frac = nearest_end_dist / road_length if road_length > 0 else np.nan
    fractions.append(frac)

acc_gdf["fraction_from_nearest_intersection"] = fractions
```

```python
nodes_gdf, edges_gdf = ox.graph_to_gdfs(G)
nodes_plot = nodes_gdf.reset_index().copy()
nodes_plot.head()
```

```python
seed_coords = [
    (center_lon - 0.004, center_lat + 0.004),  # NW
    (center_lon + 0.004, center_lat + 0.004),  # NE
    (center_lon - 0.004, center_lat - 0.004),  # SW
    (center_lon + 0.004, center_lat - 0.004),  # SE
]
```

```python
seed_nodes = []
for lon, lat in seed_coords:
    node_id = ox.distance.nearest_nodes(G, lon, lat)
    seed_nodes.append(node_id)

seed_nodes
```

```python
from networkx.algorithms.voronoi import voronoi_cells

G_u = ox.convert.to_undirected(G)

cells = voronoi_cells(G_u, seed_nodes, weight="length")
```

```python
node_to_seed = {}
for seed, node_set in cells.items():
    for n in node_set:
        node_to_seed[n] = seed

nodes_plot["cell_seed"] = nodes_plot["osmid"].map(node_to_seed)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
edges_gdf.plot(ax=ax, color="lightgray", linewidth=0.8)
nodes_plot.plot(ax=ax, column="cell_seed", cmap="tab10", markersize=8, legend=True)

# seed nodes を強調
seed_gdf = nodes_plot[nodes_plot["osmid"].isin(seed_nodes)]
seed_gdf.plot(ax=ax, color="black", markersize=50)

ax.set_title("Node-network Voronoi cells for 4 marathon seed points")
ax.set_axis_off()
plt.show()
```

```python
cell_subgraphs = {}
for seed, node_set in cells.items():
    H = G_u.subgraph(node_set).copy()
    cell_subgraphs[seed] = H
```

```python
import random
import networkx as nx

def path_length(G, path):
    total = 0
    for u, v in zip(path[:-1], path[1:]):
        edge_data = G.get_edge_data(u, v)
        if edge_data is None:
            return None
        # MultiGraph 対応: 最短lengthのedgeを使う
        min_len = min(d["length"] for d in edge_data.values())
        total += min_len
    return total

def find_approx_loop(H, start, target_length=42000, tolerance=1000, n_trials=200):
    nodes = list(H.nodes)
    if start not in H:
        return None, None

    for _ in range(n_trials):
        target = random.choice(nodes)
        if target == start:
            continue

        try:
            outward = nx.shortest_path(H, start, target, weight="length")
            outward_len = path_length(H, outward)
            if outward_len is None:
                continue

            # 往復で長さを作る
            total_len = outward_len * 2

            if abs(total_len - target_length) <= tolerance:
                loop_path = outward + outward[-2::-1]
                return loop_path, total_len

        except nx.NetworkXNoPath:
            continue

    return None, None
```

```python
results = {}

for seed in seed_nodes[:3]:
    H = cell_subgraphs[seed]
    
    if len(H.nodes) == 0:
        results[seed] = (None, None)
        continue

    start = seed
    loop_path, loop_len = find_approx_loop(H, start, target_length=42000, tolerance=1500, n_trials=500)
    results[seed] = (loop_path, loop_len)

results
```

```python
def plot_cell_and_route(H, route=None, title=""):
    fig, ax = plt.subplots(figsize=(8, 8))
    
    H_nodes, H_edges = ox.graph_to_gdfs(H)
    H_edges.plot(ax=ax, color="lightgray", linewidth=1)

    if route is not None and len(route) > 1:
        route_edges = list(zip(route[:-1], route[1:]))
        route_lines = []
        for u, v in route_edges:
            data = H.get_edge_data(u, v)
            if data is None:
                continue
            first_key = list(data.keys())[0]
            geom = data[first_key].get("geometry", None)
            if geom is None:
                # geometry がない場合は直線補間
                p1 = H.nodes[u]
                p2 = H.nodes[v]
                from shapely.geometry import LineString
                geom = LineString([(p1["x"], p1["y"]), (p2["x"], p2["y"])])
            route_lines.append(geom)

        if route_lines:
            route_gdf = gpd.GeoDataFrame(geometry=route_lines, crs="EPSG:4326")
            route_gdf.plot(ax=ax, linewidth=3)

    ax.set_title(title)
    ax.set_axis_off()
    plt.show()
```

```python
for seed, (loop_path, loop_len) in results.items():
    H = cell_subgraphs[seed]
    plot_cell_and_route(H, loop_path, title=f"Seed {seed}, loop length={loop_len}")
```

```python
all_results = {}

for seed in seed_nodes:
    H = cell_subgraphs[seed]
    if len(H.nodes) == 0:
        all_results[seed] = (None, None)
        continue

    loop_path, loop_len = find_approx_loop(H, seed, target_length=42000, tolerance=1500, n_trials=800)
    all_results[seed] = (loop_path, loop_len)

for seed, (loop_path, loop_len) in all_results.items():
    print(seed, loop_len)
```
